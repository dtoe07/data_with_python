## 1. Workout 1: Lists

#### 1. List Creation and Manipulation

* Create a list named `numbers` containing the integers from 1 to 10
* Append the number 11 to the end of the list
* Insert the number 0 at the beginning of the list
* Remove the number 5 from the list
* Reverse the order of the list

```python
import numpy as np
# Create a list named numbers containing the integers from 1 to 10
numbers = list(range(1, 11))
print("Numbers List:", numbers)

# Append the number 11 to the end of the list
numbers.append(11)
print("Appended 11 to the end:", numbers)

# insert the number 0 at the beginning of the list
numbers.insert(0, 0)
print("Inserted 0 at the beginning:", numbers)

# Remove the number 5 from the list
numbers.remove(5)
print("Removed 5:", numbers)

# Reverse the order of the list
numbers.reverse()
print("Reversed list:", numbers)
```

#### 2. List Comprehension

* Create a new list `squared_numbers` that contains the square of each number in the numbers list using list comprehension.
* Create a new list `even_numbers` that contains only the even numbers from the numbers list using list comprehension.
* Create a list `odd_squared` that contains the square of the odd numbers from the numbers list.

```python
# Create a new list squared_numbers that contains the square of each number in the numbers list using list comprehension.
squared_numbers = [x**2 for x in numbers]
print("Squared numbers:", squared_numbers)

# Create a new list even_numbers that contains only the even numbers from the numbers list using list comprehension.
even_numbers = [x for x in numbers if x % 2 == 0]
print("Even numbers:", even_numbers)

# Create a list odd_squared that contains the square of the odd numbers from the numbers list.
odd_squared = [x**2 for x in numbers if x % 2 != 0]
print("Odd squared numbers:", odd_squared)
```

#### 3. Slicing and Indexing

* Print the first three elements of the `numbers` list.
* Print the last two elements of the `numbers` list.
* Print every other element of the `numbers` list.

```python
# Print the original list
print("original list:", numbers)

# Print the first three elements of the numbers list.
print("First three elements:", numbers[:3])

# Print the last two elements of the numbers list.
print("Last two elements:", numbers[-2:])

# Print every other element of the numbers list.
print("Every other element:", numbers[::2])
```

## Workout 2: Sets

#### 1. Set Creation and Operations

* Create two sets: `set1` containing {1, 2, 3, 4, 5} and `set2` containing {4, 5, 6, 7, 8}.
* Find and print the union of `set1` and `set2`.
* Find and print the intersection of `set1` and `set2`.
* Find and print the difference of `set1` and `set2`.
* Find and print the symmetric difference of `set1` and `set2`.

```python

# Create two sets: set1 containing {1, 2, 3, 4, 5} and set2 containing {4, 5, 6, 7, 8}.
set1 = {1, 2, 3, 4, 5}
set2 = {4, 5, 6, 7, 8}

print("Set 1:", set1)
print("Set 2:", set2)

# Find and print the union of set1 and set2.
union_set = set1.union(set2)
print("Union of set1 and set2:", union_set)

# Find and print the intersection of set1 and set2.
intersection_set = set1.intersection(set2)
print("Intersection of set1 and set2:", intersection_set)

# Find and print the difference of set1 and set2.
difference_set = set1.difference(set2)
print("Difference of set1 and set2:", difference_set)

# Find and print the symmetric difference of set1 and set2.
symmetric_difference_set = set1.symmetric_difference(set2)
print("Symmetric difference of set1 and set2:", symmetric_difference_set)
```

#### 2. Set Comprehension:

* Create a set, `set3` containing the squares of numbers from 1 to 10.

```python
# Create a set, set3 containing the squares of numbers from 1 to 10.
set3 = {x**2 for x in range(1, 11)}
print("Set of squares:", set3)
```

## Workout 3: Tuples

#### 1. Tuple Creation and Unpacking:

* Create a tuple called `daily_highs` representing 60 days of temperatures (in Fahrenheit, 0 to 100 degrees)
    * Use random numbers between 40 and 90 to simulate the data
* Convert the `daily_highs` list into a tuple.

```python
# Create a tuple called daily_highs representing 60 days of temperatures (in Fahrenheit, 0 to 100 degrees) Use random numbers between 40 and 90 to simulate the data
import random
daily_highs = np.random.randint(40, 91, size=60).tolist()
print("Daily highs:", daily_highs)

# Convert the daily_highs list into a tuple.
daily_highs_tuple = tuple(daily_highs)
print("Daily highs tuple:", daily_highs_tuple)
```

#### 2. Tuple Operations

* Count the number of times the value 85 appears in the tuple.
* Try to modify the tuple. (This should cause an error)

```python
# Count the number of times the value 85 appears in the tuple.
count_85 = daily_highs_tuple.count(85)
print("Count of 85:", count_85)

# Try to modify the tuple. (This should cause an error)
try:
    daily_highs_tuple[0] = 100
except TypeError as e:
    print("Error:", e)
```

## Workout 4: Dictionaries

#### 1. Dictionary Creation and Access:

* Create a dictionary `student` with keys `name`, '`major`, and `gpa`.
* Access the value associated with the 'name' key.
* Add a new key `city`.
* Modify an existing key value pair.
* Delete a key value pair.

```python
# Create a dictionary student with keys name, 'major, and gpa.
student = {
    'name': 'Dinh Nguyen',
    'major': 'Computer Science',
    'gpa': 3.5
}
print("Student dictionary:", student)

# Access the value associated with the 'name' key.
student_name = student['name']
print("Student name:", student_name)

# Add a new key city.
student['city'] = 'Portland'
print("Student dictionary after adding city:", student)

# Modify an existing key value pair.
student['gpa'] = 3.9
print("Student dictionary after modifying GPA:", student)

# Remove the key 'major'.
del student['major']
print("Student dictionary after removing major:", student)
```

#### 2. Dictionary Comprehension:

* Create a dictionary `squared_dict` where the keys are numbers from 1 to 5 and the values are their squares.
* Create a dictionary `tuple_dict` from a list of tuples `list_of_tuplea`.

```python
# Create a dictionary squared_dict where the keys are numbers from 1 to 5 and the values are their squares.
squared_dict = {x: x**2 for x in range(1, 6)}
print("Squared dictionary:", squared_dict)

# Create a dictionary tuple_dict from a list of tuples list_of_tuplea.
list_of_tuples = [('a', 1), ('b', 2), ('c', 3)]
tuple_dict = dict(list_of_tuples)
print("Tuple dictionary:", tuple_dict)
```

#### 3. Dictionary Methods:

* Print all the keys in the `student` dictionary.
* Print all the values in the `student` dictionary.
* Print all the key-value pairs in the `student` dictionary.

```python
# Print all the keys in the student dictionary.
print("Student dictionary keys:", student.keys())

# Print all the values in the student dictionary.
print("Student dictionary values:", student.values())

# Print all the key-value pairs in the student dictionary.
print("Student dictionary items:", student.items())
```
